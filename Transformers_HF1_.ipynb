```python
# Use case #2 Automatic Speech Recognition (ASR)
```

```python
import torch

from transformers import pipeline

speech_recognizer = pipeline ("automatic-speech-recognition", model= "facebook/wav2vec2-base-960h")
```

```python
# Load an Audiodataset

from datasets import load_dataset, Audio

dataset = load_dataset("PolyAI/minds14", name= "en-US", split="train")
```

```python
dataset = dataset.cast_column("audio", Audio(sampling_rate=speech_recognizer.feature_extractor.sampling_rate))
```

```python
result = speech_recognizer(dataset[:4]["audio"])

print([d["text"] for d in result ])
```

```python
# Use case #3 Running Sentiment Classification on French text
```

```python
# Use the tags on the Hub to filter for an appropriate model. The top filtered result returns a multilingual BERT model fine tuned for sentiment analysis.

model_name= "nlptown/bert-base-multilingual-uncased-sentiment"
```

```python
# Using PYTORCH library
# using TFAutoModelForSequenceClassification and AutoTokenizer to load the pretrained model and its
# associated tokenizer

from transformers import AutoTokenizer, AutoModelForSequenceClassification

model= AutoModelForSequenceClassification.from_pretrained(model_name)

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
#using TENSORFLOW library
# using TFAutoModelForSequenceClassification and AutoTokenizer to load the pretrained model and its
# associated tokenizer


from transformers import AutoTokenizer, TFAutoModelForSequenceClassification

model= TFAutoModelForSequenceClassification.from_pretrained(model_name)

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Specify the model and tokenizer in the pipeline() and apply the classifier on French text:

classifier = pipeline("sentiment-analysis", model=model, tokenizer= tokenizer)

classifier("Nous sommes très heureux de vous présenter la bibliothèque 🤗 Transformers.")
```
